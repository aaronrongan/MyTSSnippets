```python
# 策略基类
class ETFStrategy():
#     ETFList = np.array([])
    
#     cg_VRDays=13
    
#     cg_AVDays=5
    
#     else:
        
    
    def __init__(self):
        
        self.ETFList = np.array([
#         '000012.SH', #国债ETF
        '000016.XSHG', #上证50
        '000300.XSHG',#沪深300
        '399006.XSHE', #创业板
        '000905.XSHG' #中证500
    ])
        
        #默认值
        self.cg_VRDays=21
    
        self.cg_AVDays=7
    
#         cg_enddate=datetime.datetime.today().strftime('%Y%m%d')
#         if self.cg_enddate=='':
#         self.cg_enddate=datetime.datetime.today().strftime('%Y%m%d')
            
#         print(self.cg_enddate)
        # 今日，如果是周六周日要减去，从周五开始算
        
    
    def GetTheDateTimePrice(self,index,thedate,thetime):
        #获取实时价格
        thedatetime=thedate + ' ' + thetime
        
        thedatetime=datetime.datetime.strptime(thedatetime,"%Y-%m-%d %H:%M:%S")
        
#         print('GetTheDateTimePrice the datetime:' + str(thedatetime))
        
        df_rt = get_price(index, start_date=thedatetime, end_date=thedatetime, frequency='1m') 
        return df_rt['close'][0]
    
    #获取单个指数的涨幅、均线差值比
    def GetIndexVR_AV(self,index, thedate,thetime):
        
#         if thetime='':
#             nowprice=self.GetTheDateTimePrice(index,'')
#             print('当前实时值：' + str(nowprice))
#         else:
        nowprice=self.GetTheDateTimePrice(index,thedate,thetime)         
#         print('GetIndexVR_AV_thedatetime price:' + str(nowprice))
    #     df_VR = pro.index_daily(ts_code=index, start_date=date_VRDays, end_date=enddate.strftime('%Y%m%d'))
        df_VR = get_bars(index, 
                         self.cg_VRDays, 
                         unit='1d',
                         fields=['date','close'],
                         include_now=False,
                         end_dt=datetime.datetime.strptime(thedate,'%Y-%m-%d'))
#         print(df_VR)

        df_VR_rows=df_VR.shape[0]-1
    #     print(df_VR_rows)

        # 计算当前实时涨幅
#         changepercent_VR=(df_VR['close'][df_VR_rows]-df_VR['close'][0])/df_VR['close'][0]
        
        changepercent_VR=(nowprice-df_VR['close'][0])/df_VR['close'][0]
        print('=====当前实时值相对' + str(self.cg_VRDays) + '日前涨幅:=====')
        print("%.2f%%" % (changepercent_VR * 100))

        ######################################
        #计算均线，注意这里rolling, mean, shift的用处
        df_AV =get_bars(index, 
                         self.cg_AVDays, 
                         unit='1d',
                         fields=['date','close'],
                         include_now=False,
                         end_dt=datetime.datetime.strptime(thedate,'%Y-%m-%d'))
        
        df_AV.sort_values(by='date', inplace=True, ascending=False)
#         print(df_AV)

        df_AV_rows=df_AV.shape[0]
    #     print(df_AV_rows)

#         df_AV['ma'] = df_AV['close'].rolling(self.cg_AVDays).mean().shift(self.cg_AVDays-df_AV_rows)
#         print(df_AV)

        # 计算当前实时值-M日均线差值
#         changepercent_AV=(df_AV['close'][df_AV_rows-1]-df_AV['ma'][df_AV_rows-1])/df_AV['close'][df_AV_rows-1]
        averageprice=(nowprice+df_AV['close'].sum())/(df_AV_rows+1)
#         print('包含今日+M日的均值:' + str(averageprice))
        changepercent_AV=(nowprice-averageprice)/nowprice
        print('=====当前实时值相对' + str(self.cg_AVDays) + '日均线差值百分值:=====')
        print("%.2f%%" % (changepercent_AV * 100))

    #     return VRData,AVData

    #     return changepercent_VR,changepercent_AV
        return "%.2f%%" % (changepercent_VR * 100),"%.2f%%" % (changepercent_AV * 100)

    
#     def GetBestIndex(ETFList,enddate,VRDays,AVDays):
#judgetime，用于盘中具体判断时间，比如。格式为。没有输入''则为实时时间
    def GetBestIndex(self,thedate, thetime):
        dicVR=dict()
        dicAV=dict()
        
        if thedate=='':
            thedate=datetime.datetime.today().strftime('%Y-%m-%d')
        
        if thetime=='':
            thetime=time.strftime("%H:%M:%S", time.localtime()) 

        for ETF in self.ETFList:
            print('============')
            print(ETF)
            print('============')

#             print('结束日期：' + self.cg_enddate)

            # 涨幅开始计算日期
#             date_VRDays_startdate=(datetime.datetime.strptime(self.cg_enddate,'%Y%m%d') 
#                                    - datetime.timedelta(days = 2)
#                                    -datetime.timedelta(days=self.cg_VRDays)).strftime('%Y%m%d')
#             print('涨幅开始日期：' + date_VRDays_startdate)

            # 均线开始日期(按三倍的均线日期采集参数)
#             print('self.cg_AVDays:' + str(self.cg_AVDays))
#             date_AVDays_startdate=(datetime.datetime.strptime(self.cg_enddate,'%Y%m%d')
#                                    -datetime.timedelta(days=self.cg_AVDays*3)).strftime('%Y%m%d')
#             print('均线开始日期：' + date_AVDays_startdate)

#             listVR_AV=self.GetIndexVR_AV(ETF, 
#                                          datetime.datetime.strptime(self.cg_enddate,'%Y%m%d').date()，
#                                         thetime)
#             print('GetBestIndex_thedatetime:' + thedate + ' ' + thetime)
            listVR_AV=self.GetIndexVR_AV(ETF, thedate, thetime)

#             print(listVR_AV)

            dicVR[ETF]=listVR_AV[0]
            dicAV[ETF]=listVR_AV[1]

        d_order=sorted(dicVR,key=lambda x:x[1],reverse=True)
        print('===================信号==================')
        if(float(dicAV[d_order[0]].strip('%'))>0):
            print('=======涨幅最大标的：' + str(d_order[0]) + '=====')
        else:
            
            print('=====所有标的都在均线以下，全部卖出====')
        print('=========================================')
        return 0
```
