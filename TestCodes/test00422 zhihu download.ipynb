```python
def get_list(author1):
#     url = 'https://www.zhihu.com/api/v4/columns/%s/articles?include=data[*].topics&limit=10' % author
#     url='https://zhuanlan.zhihu.com/api/columns/%s/articles?include=data%5B%2A%5D.admin_closed_comment%2Ccomment_count%2Csuggest_edit%2Cis_title_image_full_screen%2Ccan_comment%2Cupvoted_followees%2Ccan_open_tipjar%2Ccan_tip%2Cvoteup_count%2Cvoting%2Ctopics%2Creview_info%2Cauthor.is_following%2Cis_labeled%2Clabel_info&limit=10&offset=10' % author
#     url='https://zhuanlan.zhihu.com/api/columns/%s/articles?include=data[*].admin_closed_comment,comment_count,suggest_edit,is_title_image_full_screen,can_comment,Cupvoted_followees,can_open_tipjar,can_tip,voteup_count,voting,topics,review_info,author.is_following,is_labeled,label_info&limit=10&offset=10' % author1
    
    #不用后缀&limit=10&offset=10
    url='https://zhuanlan.zhihu.com/api/columns/%s/articles?include=data[*].admin_closed_comment,comment_count,suggest_edit,is_title_image_full_screen,can_comment,Cupvoted_followees,can_open_tipjar,can_tip,voteup_count,voting,topics,review_info,author.is_following,is_labeled,label_info' % author1
    article_dict = {}
    
    ###################################
    while True:
        print('fetching', url)
        try:
            resp = requests.get(url, headers=headers)
            j = resp.json()
            data = j['data']
            print(j)
        except:
            print('get list failed')

        for article in data:
            aid = article['id']
            akeys = article_dict.keys()
            if aid not in akeys:
                article_dict[aid] = article['title']

        if j['paging']['is_end']:
            break
        url = j['paging']['next']

        url=url.split('columns')
        url=url[0]+'api/columns'+url[1] 

        time.sleep(2)
    ###################################
    
    with open(HtmlSavePath + articlelistfilename, 'w',encoding='utf-8') as f:
        items = sorted(article_dict.items())
        for item in items:
            f.write('%s %s\n' % item)
```

```python
def get_html(aid, title, index):
    title = title.replace('/', '／')
    title = title.replace('\\', '＼')
    file_name = HtmlSavePath + '%03d. %s.html' % (index, title)
    
    file_name=file_name.replace('•','.')
    
    if os.path.exists(file_name):
        print(title, 'already exists.')
        return
    else:
        print('saving', title)
    try:
        url = 'https://zhuanlan.zhihu.com/p/' + aid
#         print('url is:' + url)
        html = requests.get(url, headers=headers).text
        soup = BeautifulSoup(html, 'lxml')
        content = soup.find(class_='Post-RichText').prettify()
        content = content.replace('data-actual', '')
        content = content.replace('h1>', 'h2>')
        content = re.sub(r'<noscript>.*?</noscript>', '', content)
        content = re.sub(r'src="data:image.*?"', '', content)
        content = '<!DOCTYPE html><html><head><meta charset="utf-8"></head><body><h1>%s</h1>%s</body></html>' % (
            title, content)
        with open( file_name, 'w', encoding="utf-8") as f: #,encoding="utf-8"
#             f.write(content, encoding="utf-8")
            f.write(content)
    except:
        print('get %s failed', title)
    time.sleep(2)
```

```python
def get_details():
    with open(HtmlSavePath + articlelistfilename,encoding='utf-8-sig') as f: #
        i = 1
        for line in f:
            lst = line.strip().split(' ')
            aid = lst[0]
            title = '_'.join(lst[1:])
            get_html(aid, title, i)
            i += 1
```

```python
def to_pdf(author1):
    import pdfkit
    print('exporting PDF...')
#     htmls = [HtmlSavePath + articlelistfilename]

    htmls=[]
    cover=HtmlSavePath + articlelistfilename
    
    for root, dirs, files in os.walk(HtmlSavePath):
        htmls += [HtmlSavePath + name for name in files if name.endswith(".html")]
#     print(HtmlSavePath+ author + '.pdf')
    
#     print(htmls)
    config = pdfkit.configuration(wkhtmltopdf = path_wk)
    
    if not os.path.exists(HtmlSavePath+ '知乎专栏_' +author1 + '.pdf'):
        pdfkit.from_file(sorted(htmls),HtmlSavePath+ '知乎专栏_' + author1 + '.pdf',configuration=config,options=options,cover=cover)
        return True
    else:
        print('PDF已存在')
        return False
        #     pdfkit.from_file(htmls,HtmlSavePath+ '知乎专栏_' +author + '.pdf',configuration=config,options=options)
```

```python
def clean_alltemphtmlfiles():
    for root, dirs, files in os.walk(HtmlSavePath):
        for each in files:
            if ".pdf" not in each:
                os.remove(HtmlSavePath+each)
```
