```python
def get_list(author1):

    
#     url='https://xueqiu.com/statuses/original/timeline.json?user_id=%s&page=1' % author1
    
    
#     url='https://xueqiu.com/%s/column' % author1
    
    article_dict = {}
    
    currentpagenumber=1
    ###################################
    while True:
        url='https://xueqiu.com/statuses/original/timeline.json?user_id=%s&page=%s' % (author1,currentpagenumber)
#         print(url)
#         resp = requests.get(url, headers=headers)
#         j = resp.json()
#         print(j['list'])    
#         print(j['count'])
#         print(j['maxPage'])
#         print('fetching', url)
        try:
            resp = requests.get(url, headers=headers)
            j = resp.json()
            articlelist = j['list']
#             print(j)
            articlenumber=j['count']
#             print(articlenumber)
            maxpagenumber=j['maxPage']
#             print(maxpagenumber)
        except:
            print('get list failed')

        for article in articlelist:
            aid = article['id']
#             print(aid)
            akeys = article_dict.keys()
            if aid not in akeys:
                article_dict[aid] = article['title']

#         if j['paging']['is_end']:
#             break
#         url = j['paging']['next']

#         url=url.split('columns')
#         url=url[0]+'api/columns'+url[1] 

        currentpagenumber+=1
        if currentpagenumber>maxpagenumber:
            break
            
        url='https://xueqiu.com/statuses/original/timeline.json?user_id=%s&page=%s' % (author1,currentpagenumber)
    
        time.sleep(2)
    ###################################
    
    with open(HtmlSavePath + articlelistfilename, 'w',encoding='utf-8') as f:
        items = sorted(article_dict.items())
        for item in items:
            f.write('%s %s\n' % item)
```

```python
def get_html(author,aid, title, index):
    title = title.replace('/', '／')
    title = title.replace('\\', '＼')
    file_name = HtmlSavePath + '%03d. %s.html' % (index, title)
    
    file_name=file_name.replace('•','.')
    print(file_name)
    
    if os.path.exists(file_name):
        print(title, 'already exists.')
        return
    else:
        print('saving', title)
    try:
        url = 'https://xueqiu.com/'+ author + '/' +aid
        print(url)
#         print('url is:' + url)
        html = requests.get(url, headers=headers).text
        soup = BeautifulSoup(html, 'lxml')
#         print(soup)
#         content = soup.find(class_='Post-RichText').prettify()
        content = soup.find(class_='article__bd').prettify()
#         print(content)
#         content = content.replace('data-actual', '')
        
#         content = content.replace('h1>', 'h2>')
#         content = re.sub(r'<noscript>.*?</noscript>', '', content)
        content = re.sub(r'src="data:image.*?"', '', content)
        content = '<!DOCTYPE html><html><head><meta charset="utf-8"></head><body><h1>%s</h1>%s</body></html>' % (
            title, content)
#         print(content)
        with open( file_name, 'w', encoding="utf-8") as f: 
            f.write(content)
    except:
        print('get %s failed', title)
    time.sleep(2)
```

```python
def get_details(author1):
    with open(HtmlSavePath + articlelistfilename,encoding='utf-8-sig') as f: #
        i = 1
        for line in f:
            lst = line.strip().split(' ')
            aid = lst[0]
            title = '_'.join(lst[1:])
            get_html(author1,aid, title, i)
            i += 1
```

```python
def to_pdf(author1):
    import pdfkit
    print('exporting PDF...')
#     htmls = [HtmlSavePath + articlelistfilename]

    htmls=[]
    cover=HtmlSavePath + articlelistfilename
    
    for root, dirs, files in os.walk(HtmlSavePath):
        htmls += [HtmlSavePath + name for name in files if name.endswith(".html")]
#     print(HtmlSavePath+ author + '.pdf')
    
#     print(htmls)
    config = pdfkit.configuration(wkhtmltopdf = path_wk)
    
    if not os.path.exists(HtmlSavePath+ '雪球专栏_' +author1 + '.pdf'):
        pdfkit.from_file(sorted(htmls),HtmlSavePath+ '雪球专栏_' + author1 + '.pdf',configuration=config,options=options,cover=cover)
        return True
    else:
        print('PDF已存在')
        return False
        #     pdfkit.from_file(htmls,HtmlSavePath+ '知乎专栏_' +author + '.pdf',configuration=config,options=options)
```

```python
def clean_alltemphtmlfiles():
    for root, dirs, files in os.walk(HtmlSavePath):
        for each in files:
            if ".pdf" not in each:
                os.remove(HtmlSavePath+each)
```
