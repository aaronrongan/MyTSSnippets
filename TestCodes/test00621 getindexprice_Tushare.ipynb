```python
#获取单个指数的涨幅、均线差值比
def GetIndexVR_AV(index, enddate, date_VRDays, date_AVDays_startdate):
    
    df_VR = pro.index_daily(ts_code=index, start_date=date_VRDays, end_date=enddate.strftime('%Y%m%d'))
#     print(df_VR)
    
    df_VR_rows=df_VR.shape[0]-1
#     print(df_VR_rows)

    # 计算涨幅
    changepercent_VR=(df_VR['close'][0]-df_VR['close'][df_VR_rows])/df_VR['close'][df_VR_rows]
#     print(changepercent_VR)
    
    ######################################
    #计算均线，注意这里rolling, mean, shift的用处
    df_AV = pro.index_daily(ts_code=index, start_date=date_AVDays_startdate, end_date=enddate.strftime('%Y%m%d'))
    df_AV.sort_values(by='trade_date', inplace=True, ascending=False)
#     print(df_AV)

    df_AV_rows=df_AV.shape[0]
#     print(df_AV_rows)
    
    df_AV['ma'] = df_AV['close'].rolling(g_AVDays).mean().shift(g_AVDays-df_AV_rows)
#     print(df_AV)

    # 计算当日值-均线差值
    changepercent_AV=(df_AV['close'][0]-df_AV['ma'][0])/df_AV['close'][0]
#     print(changepercent_AV)
    
#     return VRData,AVData
    
#     return changepercent_VR,changepercent_AV
    return "%.2f%%" % (changepercent_VR * 100),"%.2f%%" % (changepercent_AV * 100)
    
```

```python
#比较几个指数的涨幅、均线差值比，返回指数
def GetBestIndex(ETFList,enddate,VRDays,AVDays):
    
    dicVR=dict()
    dicAV=dict()
    
    for ETF in ETFList:
        print('============')
        print(ETF)
        print('============')
        
        print('结束日期：' + enddate)
    
        # 涨幅开始计算日期
        date_VRDays=(datetime.datetime.strptime(enddate,'%Y%m%d') - datetime.timedelta(days = 2)-datetime.timedelta(days=VRDays)).strftime('%Y%m%d')
        print('涨幅开始计算日期：' + date_VRDays)

        # 均线开始日期(按三倍的均线日期采集参数)
        date_AVDays_startdate=(datetime.datetime.strptime(enddate,'%Y%m%d')-datetime.timedelta(days=AVDays*3)).strftime('%Y%m%d')
        print('均线开始日期：' + date_AVDays_startdate)
        
#         print(GetIndexVR_AV(ETF, datetime.datetime.strptime(date_today,'%Y%m%d').date(),date_VRDays,date_AVDays_startdate))
        
        listVR_AV=GetIndexVR_AV(ETF, datetime.datetime.strptime(date_today,'%Y%m%d').date(),date_VRDays,date_AVDays_startdate)
        
#         print(listVR_AV)
        
        dicVR[ETF]=listVR_AV[0]
        dicAV[ETF]=listVR_AV[1]
    
#     print(dicVR)
#     print(dicAV)
    d_order=sorted(dicVR,key=lambda x:x[1],reverse=True)
    if(float(dicAV[d_order[0]].strip('%'))>0):
        print(d_order[0])
    else:
        print('所有标的都在均线以下')
    return 0
```
